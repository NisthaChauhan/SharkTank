```python
sns.set_style('whitegrid')
sns.set_theme('poster')
#METHOD 1
plt.figure(figsize=(20,5))
x= df['Yearly Revenue']
y = df['Monthly Sales']
sns.regplot(x=x, y=y)
plt.xlabel('Yearly Revenue')
plt.ylabel('Monthly Sales')
plt.title('Regression Plot of Yearly Revenue vs Monthly Sales')
plt.show()

#METHOD 2
plt.figure(figsize=(20,5))
x = df["Original Offered Equity"]
y = df["Valuation Requested"]
s,i,r,p,std_err = stats.linregress(x,y)
def func(x):
    return s*x + i
o = list(map(func,x))
plt.xlabel("Original Offered Equity")
plt.ylabel("Valuation Requested")
plt.scatter(x,y)
plt.plot(x,o)
plt.show() 
''' 
#Regression can be plot for : 

    Yearly Revenue vs Monthly Sales
    Gross Margin vs Net Margin
    EBITDA vs Total Deal Amount
    Number of Presenters vs Total Deal Amount
    Male Presenters vs Female Presenters
    Total Deal Equity vs Total Deal Debt
    Namita Investment Amount vs Namita Investment Equity
    Vineeta Investment Amount vs Vineeta Investment Equity
    Anupam Investment Amount vs Anupam Investment Equity
    Aman Investment Amount vs Aman Investment Equity
'''
```
